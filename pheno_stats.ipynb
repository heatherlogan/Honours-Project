```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

```

```python
# full paper phenotype distribution

data_path = os.path.join(os.getcwd(), 'files/stats/full_corpus_counts/',
                         'fullcorpus_pheno_stats.csv')

full_df = pd.read_csv(data_path)
```

```python
group_keys = full_df.groupby('Group').size().keys()

print(group_keys)
```

```python
group_colours = [
    ('Perinatal History', '') ,
    ('Adaptive Life Skills', '') ,
    ('Cognitive Abilties', '') ,
    ('Comorbidities', '') ,
    ('Complaints and indications', '') ,
    ('Diagnosis', '') ,
    ('Emotional Traits', '') ,
    ('Executive Function', '') ,
    ('Exposures', '') ,
    ('Interpersonal Interactions', '') ,
    ('Language Ability', '') ,
    ('Motor Skills', '') ,
    ('Recognition of Social Norms', '') ,
    ('Stereotyped Restricted and Repetitive Behaviours', '') ,

]
```

```python
# 0=medical, 1=personal, 2=social
group_tags = {
    'Perinatal History': 0 ,
    'Adaptive Life Skills':2 ,
    'Cognitive Abilties':1 ,
    'Comorbidities':0,
    'Complaints and indications':0 ,
    'Diagnosis': 0,
    'Personal Traits': 1,
    'Emotional Traits':1 ,
    'Executive Function':1 ,
    'Exposures':0,
    'Cognitive Abilities':1,
    'Interpersonal Interactions':2 ,
    'Language Ability':1,
    'Motor Skills':1 ,
    'Recognition of Social Norms':2,
    'Stereotyped Restricted and Repetitive Behaviours':2,
    'Social Competence':2,
    'Medical History':0
}

print(group_colours)
```

```python
colors = {0:'#7FB3D5', 1:'#BB8FCE', 2:'#F5CBA7'}
```

```python
idx_list = []
for col in full_df.columns:
    if col == "Group":
        for i, row_value in full_df[col].iteritems():
            g = (group_tags[row_value])
            idx_list.append(g)
print(len(idx_list))

full_df['Group_idx'] = idx_list
```

```python
sorted_full = full_df.sort_values(['Group_idx', 'Group', 'RelevantCount'], 
                                  ascending=[True,True, False])

sorted_full
```

```python
colour_list = []
for col in full_df.columns:
    if col == "Group_idx":
        for i, row_value in full_df[col].iteritems():
            colour_list.append(colors[row_value])
            
print(colour_list)
```

```python
fig, ax = plt.subplots(figsize=(20,50))

plt.scatter(sorted_full['RelevantCount'], sorted_full['Label'], color=colour_list)
```

```python
groups = sorted_full.groupby("Group").sum().reset_index()


idx_list = [] 
for col in groups.columns:
    if col == "Group":
        for i, row_value in groups[col].iteritems():
            g = (group_tags[row_value])
            idx_list.append(g)

groups["Group Idx"] = idx_list


sorted_full
```

```python
sort_groups = groups.sort_values(["Group Idx", "Count"], ascending=[False,False])

colour_list = []
for col in sort_groups.columns:
    if col == "Group Idx":
        for i, row_value in sort_groups[col].iteritems():
            colour_list.append(colors[row_value])
            
print(colour_list)

```

```python

fig, ax = plt.subplots(figsize=(30,20))


plt.figure(1)
plt.subplot(121)
plt.barh(sort_groups['Group'],sort_groups['Count'], color=colour_list)
plt.xlabel("Documents appeared in.")
plt.rcParams.update({'font.size': 22})



plt.subplot(122)
plt.barh(sort_groups['Group'],sort_groups['RelevantCount'], color=colour_list)
plt.yticks([])
plt.xlabel("Number of Documents annotation is important to.")
plt.rcParams.update({'font.size': 22})



plt.tight_layout()
plt.show()
```

```python
comorbid_labels = []

for col in comorbids.columns:
    if col == "Label":
        for i, row_value in comorbids[col].iteritems():
            for label, symplist in comorbid_groups.items():
                if row_value in symplist:
                    comorbid_labels.append(label)
                    
comorbids["labels"] = comorbid_labels
                    
comorbids = comorbids.sort_values(['labels', 'RelevantCount'], ascending=[False,False])

comorbids
```

```python
com_colours = {"Autoimmune":'#0B5345', 
        "Bacterial":'#F5CBA7',
         "Cardiovascular":'#E74C3C',
         "Congenital Abnormalities":'#5499c7',
         "Ear Diseases":'#1A5276',
         "Endocrine": '#E67E22', 
         "Gastrointestinal": '#BB8FCE', 
         "Mental Disorder": '#16A085',
         "Nervous System": '#884EA0',
         "Urogenital":'#1f618d',
         "Virus":'#62b3c4'}
```

```python
colour_list_com = []
for col in comorbids.columns:
    if col == "labels":
        for i, row_value in comorbids[col].iteritems():
            colour_list_com.append(com_colours[row_value])
            
print(colour_list_com)
```

```python
import matplotlib.patches as mpatches

patches = []
for l, c in com_colours.items():
    patches.append(mpatches.Patch(color=c, label=l))
    
```
